```python
var=movies.iloc[1]['genres']
```

```python
def genres_data(data):
    lst=[]
    for i in ast.literal_eval(var):
        lst.append(i['name'])
    return lst
```

```python
def actors(data):
    lst=[]
    for i in ast.literal_eval(data):
        if i['order'] < 5:
            lst.append(i['name'])
    return lst
```

```python
actors(movies.iloc[1]['cast'])
```

```python
movies['cast'].apply(actors)
```

```python
movies['cast']=movies['cast'].apply(actors)
```

```python
def actor(data):
    counter=0
    lst=[]
    for i in ast.literal_eval(data):
        if counter<=5:
            lst.append(i['name'])
            counter+=1
    return lst
```

```python
movies['cast']=movies['cast'].apply(actors)
```

```python
def space_remover(data):
    lst=[]
    for i in data:
        lst.append(i.replace(" ",""))
    return lst
```

```python
import ast
```

```python
def fetch_director(data):
    lst=[]
    for i in ast.literal_eval(data):
        if i['job']== "Director":
            lst.append(i["name"])
    return lst
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['crew'].apply(space_remover)
```

```python
movies['crew'] = movies['crew'].apply(space_remover)
movies
```

```python
movies['cast'] = movies['cast'].apply(space_remover)
movies['crew'] = movies['crew'].apply(space_remover)
movies['genres']=movies['genres'].apply(space_remover)
movies['keywords']=movies['keywords'].apply(space_remover)
```
